```python
locations = ["Bengaluru", "Mysuru", "Mangaluru", "Hubli", "Belagavi", "Tumakur", "Udupi", "Shimoga", "Bellary", "Davanagere", "Hassan","Bidar", "Raichur", "Haveri", "Yadgiri", "Chikkamagaluru", "Mandya", "Kolar", "Koppal"]
categories = ["Food loss", "Waste in general", "Monetary and economic impact", "Others"]
activities = ['Households', 'Combined: Eurostat proxy for manufacturing of food products - includes tobacco',
              'Agricultural production', 'Combined: Eurostat proxy for services',
              'Combined : Post-production and Pre-consumption', 'Total',
              'Wholesale & Retail', 'Retail', 'Combined : pre-retail',
              'Food Industry', 'Restaurants',
              'Combined: Households and food services activities',
              'Combined: Retail + food services + households', 'Food Services',
              'Contextual data', 'Combined: food service to households',
              'Municipal Waste', 'Wholesale', 'School cantines',
              'Combined : Post-production', 'Combined: Pre-consumption',
              'Combined: Retail and Food services', 'Grocery stores', 'Producers',
              'Combined: retail to households', 'Manufacturing of Food products',
              'Industrial kitchens', 'Large scale consumers', 'Combined', 'Trade',
              'Transportation', 'Storage', 'Large restaurants',
              'Mass feeding facilities', 'Manufacturing']
units = ["Tonnes", "%"]
variables = ["Unnecessary edible food waste", "Food Waste - Non-edible", "Food Loss & Waste in value of food production",
             "Food Loss from restaurants and cafeterias", "Food Waste", "Food Loss", "Food Waste - Fruit",
             "Food Waste - Vegetables", "Food Waste - Rice", "Food Waste - Meat", "Food Waste - Poultry",
             "Food Loss from harvest", "Animal food waste", "Animal and mixed food waste; vegetal waste",
             "Household food waste", "Food Loss - Fresh Fruit"]
times = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
values = np.random.randint(100, 5000, size=300)  # Random values for the 'Value' column
```

```python
data = []
```

```python
for _ in range(300):
    row = {
        "Location": np.random.choice(locations),
        "Category": np.random.choice(categories),
        "Activity": np.random.choice(activities),
        "Unit": np.random.choice(units),
        "Variable": np.random.choice(variables),
        "Time": np.random.choice(times),
        "Value": np.random.choice(values),
        "Flags": np.nan
    }
    data.append(row)
```

```python
def build_model():
  model = keras.Sequential([
      layers.Dense(64, activation = tf.nn.relu, input_shape=[len(train_dataset.keys())]),
      layers.Dense(64, activation = tf.nn.relu),
      layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer = optimizer,
                metrics=['mae', 'mse'])

  return model
```
